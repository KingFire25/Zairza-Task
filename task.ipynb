#### Create an array of all the odd integers from 100 to 200

```python
a=[]
for i in range(100,200):
    if(i%2!=0):
        a.append(i)
a=np.array(a)
```

#### Create a 3x3 identity matrix

```python
b=[]
for i in range(3):
    a=[]
    for j in range(3):
        if i==j:
            a.append(1)
        else:
            a.append(0)
    b.append(a)
b=np.array(b)
```
